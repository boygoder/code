```python
import matplotlib.pyplot as plt

sizes = [10000,20000,40000,80000,160000];
```

```python
proportion = 0.01
quickRunTime = [4,26,72,122,280]
heapRunTime = [68,155,353,689,1424]
plt.title("Orderliness = 0.01")
plt.xlabel('size')
plt.ylabel('time');
plt.plot(sizes, quickRunTime, marker='o', markersize=3)  
plt.plot(sizes, heapRunTime, marker='o', markersize=3)  
plt.legend(["quickSort", "heapSort"])  # 设置折线名称
plt.show()
```
