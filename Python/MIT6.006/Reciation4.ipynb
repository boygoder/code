```python
class DirectAccessArray:
    def __init__(self,u):
      self.A = [None]*u;
    def find(self,k):
        return self.A[k];
    def insert(self,x):
        self.A[x] = x;  
    def delete(self,k):
        self.A[k] = None;
    def find_next(self,k):    
        for i in range(k,len(self.A)):
            if self.A[i] is not None:
                return self.A[i]
    def find_max(self):
        for i in range(len(self.A)-1,-1,-1):
            if self.A[i] is not None:
                return self.A[i]
    def delete_max(self):
        for i in range(len(self.A)-1,-1,-1):
            x = self.A[i]
            if x is not None:
                self.A[i] = None
                return x
```

```python
class Hash_Table_Set:
    def __init__(self,r = 200):
        self.chain_set = Set_from_Seq(Linked_List_Seq)
        self.A =[]
        self.size = 0
        self.r = r; #100/self.r = 填充比例
        self.p = 2**31-1
        self.a = math.randint(1,self.p-1)
        self.b = 0
        self._compute_bounds()
        self._resize(0)
    def __len__(self):
        return self.size();
    def __iter__(self):
        for X in self.A:
            yield from X
    def build(self,X):
        for x in X:
            self.insert(x)
    def _hash(self,k,m):
        return (((self.a * k)% self.p) %m);
    def _compute_bounds(self):
        self.upper = len(self.A)
        self.lower = len(self.A)*100 *100 //(self.r*self.r);
    def _resize(self,n):
        if(self.lower >= n) or (self.upper <= n):
            f = self.r // 100
            if self.r % 100: f+=1; #f = ceil(r/100);
            m = max(n,1)*f;
            A = [self.chain_set() for _ in range(m)]
            for x in self:
                h = self._hash(x.key,m)
                A[h].insert(x)
            self.A = A
            self._compute_bounds()
    def find(self,k):
        h = self._hash(k,len(self.A))
        return self.A[h].find(k);
    def insert(self,x):
        self._resize(self.size+1)
        h = self._hash(x,len(self.A))
        added = self.A[h].insert(x)
        if added: self.size+=1
        return added
    def delete(self,k):
        assert(len(self)>0)
        h = self._hash(k,len(self.A))
        x = self.A[h].delete(k)
        self.size -=1
        self._resize(self.size)
        return x
    def find_min(self):
        out = None
        for x in self:
            if (out is None) or (x.key < out.key):
                out = x
        return out
    def find_max(self):
        out = None
        for x in self:
            if(out is None) or (x.key > out.key):
                out  = x
        return None
    def find_next(self,k):
        out = None
        for x in self:
            if x.key > k:
                if(out is None) or (x.key < out.key):
                    out = x
        return out
    def find_prev(self,k):
        out = None
        for x in self:
            if x.key < k:
                if(out is None) or (x.key > out.key):
                    out  = x
        return out
    def iter_order(self):
        x = self.find_min()
        while x:
            yield from x
            x = self.find_next(x.key)
```

```python
def Set_from_Seq(seq):
    class set_from_seq:
        def __init__(self):
            self.S = seq()
        def __len__(self):
            return len(self.S)
        def __iter__(self):
            yield from self.S 
        def build(self,X):
            self.S.build(X)
        def find(self,k):
            for x in self.S:
                if(x.key==k):
                    return x
            return None
        def insert(self,x):
            for i in range(len(self.S)):
                if(self.S.get_at(i) == x):
                    self.S.set_at(i, x.key)
                    return None
            self.S.insert_last(x)
        def delete(self,k):
            out = None
            for i in range(len(self.S)):
                if(self.S.get_at(i).key == k):
                    out = self.S.get_at(i)
                    break;
            return out;
        def iter_ord(self): 
            item = self.find_min()
            while(item):
                yield item
                item = self.find_next(item.key)
        def find_min(self):
            out = None
            for i in range(len(self.S)):
                now = self.S.get_at(i)
                if (out == None) or ( now.key< out.key):
                        out = now
            return out
        def find_max(self):
            out = None
            for i in range(len(self.S)):
                now = self.S.get_at(i)
                if (out == None) or ( now.key > out.key):
                    out = now
            return out
        def find_next(self,k):
            out = None
            for i in range(len(self.S)):
                now = self.S.get_at(i)
                if(now.key > k):
                    if(out == None) or ( now.key > out.key):
                        out = now
            return out
        def find_prev(self,k):
            out = None
            for i in range(len(self.S)):
                now = self.S.get_at(i)
                if(now.key < k):
                    if(out == None) or ( now.key < out.key):
                        out = now
            return out  
    return set_from_seq   
```
