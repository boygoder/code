```python
def height(A):
    if A:
        return A.height
    else:
        return -1

class Binary_Node:

    
    def __init__(A,x):
        A.item = x
        A.left = None
        A.right = None
        A.parent = None
        A.subtree_update()

    def subtree_update(A):
        A.height =  1 + max(height(A.left),height(A.right));

    def skew(A):
        return height(A.right) -  height(A.left)

    def subtree_iter(A):
        if A.left:
            yield from A.left.subtree_iter()
        yield A
        if A.right:
            yield from A.right.subtree_iter()
    def subtree_first(A):
        if A.left:
            return A.left.subtree_first()
        else:
            return A
    def subtree_last(A):
        if A.right:
            return A.right.subtree_last()
        else:
            return A
    def successor(A):
        if A.right:
            return A.right.subtree_first();
        else:
            while (A.parent) and (A == A.parant.right):
                A = A.parent
            return A.parent;
    def predecessor(A):
        if A.left:
            return A.left.subtree_last();
        else:
            while (A.parent) and (A == A.parent.left):
                A = A.parent
            return A.parent;            
    def subtree_insert_before(A,B):
        if(A.left):
            pNode = A.left.subtree_last();
            B.parent = pNode;
            pNode.right = B;
        else :
            B.parent = A;
            A.left = B;
        A.maintain();
    def subtree_insert_after(A,B):
        if(A.right):
            pNode = A.right.subtree_first();
            B.parent = pNode;
            pNode.left = B;
        else:
            B.parent = A;
            A.right = B; 
        A.maintain();
    def subtree_delete(A):
        if A.left or A.right: # A is not a leaf
            if A.left:
                B = A.predecessor();
            else:
                B = A.successor();
            A.item,B.item = B.item,A.item;
            return B.subtree_delete();
        
        if A.parent:#A is a leaf
            if A == A.parent.left:
                A.parent.left = None
            else:
                A.parent.right = None
            A.parent.maintain();
        return A
    def subtree_rotate_right(D):
        assert D.left
        B,E = D.left,D.right
        A,C = B.left,B.right
        D,B = B,D
        D.item,B.item = B.item,D.item
        B.left,B.right = A,D
        D.left,D.right = C,E
        if A is not None:
            A.parent = B
        if E is not None:
            E.parent = D
        B.subtree_update()
        D.subtree_update()
    def subtree_rotate_left(B):
        assert B.right
        A,D = B.left,B.right
        C,E = D.left,D.right
        D,B = B,D
        D.item,B.item = B.item,D.item
        D.left,D.right = B,E
        B.left,B.right = A,C
        if A is not None:
            A.parent = B
        if E is not None:
            E.parent = D
        B.subtree_update()
        D.subtree_update()
    def  reblance(A):
        if A.skew() == 2:
            if A.right.skew() < 0:
                A.right.subtree_rotate_right();
            A.subtree_rotate_left()
        elif A.skew() == -2:
            if A.left.skew() > 0:
                A.left.subtree_rotate_left();
            A.subtree_rotate_right();
    def maintain(A):
        A.reblance()
        A.subtree_update()
        if A.parent:
            A.parent.maintain();            
```

```python
class Binary_Tree:
    def __init__(T,Node_Type = Binary_Node):
        T.root = None
        T.size  = 0
        T.Node_Type = Node_Type
    
    def __len__(T):
        return T.size;
    
    def __iter__(T):
        if T.root:
            for A in T.root.subtree_iter():
                yield  A.item
    # 将数组转换为按下标次序插入的二叉树，且高度为logN
    def build(self,X):
        A = [x  for x in X]
        def build_subtree(A,i,j):
            c = (i+j) // 2
            root = self.Node_Type(A[c])
            if i < c:
                root.left = build_subtree(A,i,c-1)
                root.left.parent = root
            if c < j:
                root.right = build_subtree(A,c+1,j)
                root.right.parent = root
            return root
        self.root = build_subtree(A,0,len(A)-1)
```

```python
class BST_Node(Binary_Node):
    def subtree_find(A,k):
        if k < A.item.key:
            if A.left:
                return A.left.subtree_find(k)
        elif k > A.item.key:
            if A.right:
                return A.right.subtree_find(k)
        else:
            return A
        return None
    def subtree_find_next(A,k):
        if A.item.key <= k:
            if A.right:
                return A.right.subtree_find_next(k)
            else:
                return None
        else:
            if A.left:#看看有没有更接近k的
                B = A.left.subtree_find_next(k)
                if B:
                    return B
            else:#没有,A就是最接近k的
                return A
        return A
    def subtree_find_prev(A,k):
        if A.item.key >= k:
            if A.left:
                return A.left.subtree_find_prev(k)
            else:
                return None
        else:
            if A.right:#看看有没有更接近k的
                B = A.right.subtree_find_prev(k)
                if B:
                    return B
            else:#没有,A就是最接近k的
                return A
        return A
    def subtree_insert(A,B):
        if B.item.key < A.item.key:
            if A.left:
                A.left.subtree_insert(B)
            else:
                A.subtree_insert_before(B)
        elif B.item.key > A.item.key:
            if A.right:
                A.right.subtree_insert(B)
            else:
                A.subtree_insert_after(B)
        else:
            A.item = B.item


```

```python
def Partition(A,p,r):
    x = A[r]
    smallEnd = p-1;
    for bigEnd in range(p,r):
        if(A[bigEnd] <= x):#bigEnd每遇见一个小的，就交换到前边去
            smallEnd +=1
            A[smallEnd],A[bigEnd] = A[bigEnd],A[smallEnd];
    A[smallEnd+1],A[r] = A[r],A[smallEnd+1]
    return smallEnd+1;
def QuickSort(A,p,r):
    if p < r:
        q = Partition(A,p,r)
        QuickSort(A,p,q-1)
        QuickSort(A,q+1,r)
```
